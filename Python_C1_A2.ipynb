## 1 - Limite e Continuidade


### 1.1 - Definição

De forma intuitiva, quando falamos da continuidade de uma função $f$ em algum ponto $p$, estamos dizendo que essa função não apresenta "saltos" no seu gráfico.


Nas figuras abaixo temos dois exemplos de funções:
- a do lado esquerdo é dita contínua em $p$;
- a do lado direito é dita descontínua em $p$.

continuidade.PNG

Retirada do livro Guidorizzi, H. Um curso de cálculo. Vol I.

**Formalmente temos:**

---

1. **Limite**: O limite de uma função $f$ existe se, e só se, $\forall ϵ > 0, \exists δ > 0 $, tal que, se $|x-p|< \delta \Rightarrow |f(x)-f(p)|< \epsilon .$

  Notação:
  
  $$ \lim_{x \rightarrow p} f(x) = L.$$

  ---

2. **Continuidade**: Um função $f$ é dita contínua em um ponto $p$, se o limite dessa função existe e é o valor da função no ponto. Ou seja, $L=f(p)$.

  Notação: $$ \lim_{x \rightarrow p} f(x) = f(p).$$

derivada.PNG

Retirada do livro Guidorizzi, H. Um curso de cálculo. Vol I.

Seja $I$ um intervalo aberto não-vazio e seja
$f : I \to \mathbb{R}$, $y = f(x)$,
uma função de $I$ em $\mathbb{R}$.
Diz-se que a função $f(x)$ é derivável no ponto
$p \in I$ se existir o seguinte limite:

$$f'(p) = \lim_{x \rightarrow p}  \frac{f(x) - f(p)}{x - p},$$


Essa expressão representa o quanto a função $f$ varia, a medida que a diferença $x - p \rightarrow 0$. O que chamamos de **taxa de variação** ou **derivada** de f no ponto p.

**Soma de Riemann**

Graficamente, quando tomamos o produto $f(c_i) (x_i - x_{i-1}) $, obtemos áreas retagulares que, **quanto menor**  for o subintervalo $\Delta x_i$, **melhor será** a aproximação do valor da área abaixo da curva gerada por $f$.

O valor
$$ \sum_{i=1} ^{n} f(c_i) \Delta x_i   $$

denomina-se **soma de Riemann da função f**.

Nessa situações é interessante contar com um método alternativo do cálculo da integral: **uma aproximação numérica por somas parciais.**

---

Dada a função

$$v(t) =t^2  cos \left( \sqrt{\dfrac{k}{t^2 + k^3}} \right)  $$

com $k>0$ e $t \in [0, T]$, (**task**) calcule área sobre a curva através da soma de Riemann

$$
\int_a ^b f(x) \ dx =  \lim_{\max \Delta x_i \to 0} \sum_{i=1}^{n} f(c_i)\,\Delta x_i = L.
$$

```python
#EXECUTE ESSE CÓDIGO MUDANDO APENAS OS ARGUMENTOS "a", "b" E "n"

import numpy as np
from numpy import sum, linspace

# definindo a partição do intervalo
n = 1000
a = 0 #extremo inferior do intervalo
b = 30 #extremo superior do intervalo
t = np.linspace(a, b, n)   #verifique que o intervalo foi dividido em "n" partes


# definindo os pontos ci (nesse caso) médios no intervalo
ci = np.zeros(len(t)-1)

for i in range(len(t)-1):
  ci[i] = (t[i+1]+t[i]) / 2             # para riemann-stielges (1-p)*t[i+1]+ p* t[i]
```

```python
# EXECUTE ESSE CÓDIGO MUDANÇA A EXPRESSÃO DA FUNÇÃO PARA CADA ITEM DO EXERCÍCIO. CONSIDERE OS "ci" COMO "x" DA FUNÇÃO

# a função falculada em ci
k = 2
f =  ci**2 * np.cos( np.sqrt( k/(ci**2 + k**3)    )  )
```

```python
# EXECUTE O CÓDIGO SEM MUDAR NADA!!

# vetor nulo
somatorio = np.zeros(len(t)-1)


# a soma de riemann
for i in range(len(t)-1):
  somatorio[i] = f[i] * (t[i+1]-t[i])

rieman = sum(somatorio)
float(rieman.round(5))
```
