### Definições

De forma intuitiva, quando falamos da continuidade de uma função $f$ em algum ponto $p$, estamos dizendo que essa função não apresenta "saltos" no seu gráfico.


Nas figuras abaixo temos dois exemplos de funções:
- a do lado esquerdo é dita contínua em $p$;
- a do lado direito é dita descontínua em $p$.

continuidade.PNG

Retirada do livro Guidorizzi, H. Um curso de cálculo. Vol I.

Vamos para algumas definições mais formais.

1. **Limite**: O limite de uma função $f$ existe se, e só se, $\forall ϵ > 0, \exists δ > 0 $, tal que, se $|x-p|< \delta \Rightarrow |f(x)-f(p)|< \epsilon .$

  Notação: $ \lim_{x \rightarrow p} f(x) = L$.

\\

2. **Continuidade**: Um função $f$ é dita contínua em um ponto $p$, se o limite dessa função existe e é o valor da função no ponto.

  Notação: $ \lim_{x \rightarrow p} f(x) = f(p)$.

## 2 - Derivação

### Definições

Considere uma função $f$ que passa por um ponto $p$ fixo, como na figura abaixo. Caso tomemos um outro ponto $x$ e o façamos se aproximar de $p$, o que acontece?

R: Encotramos o limite de $f$ quando $x$ tende a $p$.

derivada.PNG

Retirada do livro Guidorizzi, H. Um curso de cálculo. Vol I.

Mas e se estivermos interessados em encontrar equação da reta que conecta esses dois pontos? Isso é, a equação

$y = f(p) + m(x-p),$

onde $m$ é o coeficiente angular dessa reta.

\\

O que acontece, é que, conforme $x$ tende para $p$, mais e mais essa reta $y$ tende para a reta tangente a $f(x)$ no ponto $p$.

Ou seja, se tomarmos

$m = \lim_{x \rightarrow p}  \frac{f(x) - f(p)}{x - p},$

obtemos a equação da reta tangente a $p$.

Essa expressão representa o quanto a função $f$ varia, a medida que a diferença $x - p \rightarrow 0$. O que chamamos de **taxa de variação** ou **derivada** de f no ponto p.

### Prática

Podemos verificar se as funções que aprendemos a declarar com o Sympy, possuem ou não derivada.

```python
import sympy as sp
from sympy import Symbol, Function

x, f, g = Symbol('x'), Function('f'), Function('g')
```

O método "diff" possibilita o cálculo da expressão geral da derivada de um função real, passando como argumento a função a ser derivada e a variável, tantas vezes quanto se quiser derivar.
\\

**Exemplo**: Declare as funções $f(x) = x^3 + x^2 + x - 2$ e $g(x) = ln(x^2)$ e use o comando "sp.diff(função, variável)" para achar a primeira derivada de cada função, isso é, $\frac{df}{dx}$ e  $\frac{dg}{dx}$.

Encontre $\frac{df}{dx}$:

```python
sp.diff(f, x)
```

Encontre $\frac{dg}{dx}$:

Encontre $\frac{d^n f}{dx^n }$ ou  $\frac{d^n g}{dx^n }$, para $n \in \mathbb{N}$:

```python
n = 3
# n é numero inteiro que indica quantas vezes f será derivada

sp.diff(g, x, n)
```

Analise o gráfico das derivadas e das funções e verifique que **derivação (diferenciabilidade) implica em continuidade**.

```python
dif_f = sp.diff(f, x)    #definir uma outra função a partir da derivada
dif_g = sp.diff(g, x)
label = ['derivada', 'funcao']   #definir os rótulos das funções
```

```python
sp.plot(dif_f, f, legend=True, label=label)
```

```python
sp.plot(dif_g, g, legend=True, label=label)
```

## 3 - Integração

### Definições

Considere um subconjunto $I \subset \mathbb{R}$, tal que $I = (a, b)$, onde $a < b$.

Além disso, considere uma partição $P$ desse intervalo de modo que

$$P:= a=x_0 < x_1 < x_2 < … < x_{n-1} < x_n = b. $$

Para cada índice $i$ ($i = 1, 2, 3, …, n$), tomamos pontos $c_i$ dos subintervalos $ [x_{i − 1}, x_i]$, escolhidos arbitrariamente, como mostrado abaixo.


particao.PNG

Retirada do livro Guidorizzi, H. Um curso de cálculo. Vol I.

Considere também uma função $f$, avaliada em cada ponto $c_i$, ou seja, $f(c_i)$. Visualmente, o que temos são elementos da imagem da função que correspondem a pontos de "altura" relativa a curva gerada pela função $f$. Observe abaixo:

riemann_baixa.jpg

Retirada do livro Carvalhaes, C. G; et al. Cálculo Diferencial e Integral de Funções de uma variável com Python.

**Soma de Riemann**

Graficamente, quando tomamos o produto $f(c_i) (x_i - x_{i-1}) $, obtemos áreas retagulares que, **quanto menor**  for o subintervalo $\Delta x_i$, **melhor será** a aproximação do valor da área abaixo da curva gerada por $f$.

O valor
$$ \sum_{i=1} ^{n} f(c_i) \Delta x_i   $$

denomina-se soma de Riemann de f.

**Integral (de Riemann)**

O que chamamos de integral (segundo Riemann) é o número $L$ (se existir tal número), encontrado quando fazemos o valor máximo dos subintervalos $\Delta x_i$ tenderem a zero. Ou seja,

```python
import numpy as np
from numpy import sum, linspace

# definindo a partição do intervalo
n = 100
x = linspace(-2, 3, n)   #verifique que o intervalo foi dividido em "n" partes


# definindo os pontos ci (nesse caso) médios no intervalo
ci = np.zeros(len(x)-1)

for i in range(len(x)-1):
  ci[i] =  (x[i+1]+x[i]) / 2

# a função falculada em ci, a função é f(x) = x**2
f = (ci**4)/4 - ci**3 - 2*ci**2 + 3
```

```python
# a soma de riemann
somatorio = np.zeros(len(x)-1)

for i in range(len(x)-1):
  somatorio[i] = f[i] * (x[i+1]-x[i])

rieman = sum(somatorio)
rieman
```

```python
# verificando se a soma é uma boa aproximação, usando as ferramentas padrões do Sympy
x = sp.Symbol('x')
f = sp.Function('f')

f =  (x**4)/4 - x**3 - 2*x**2 + 3
sp.integrate(f, (x, -2, 3)).evalf()  #evalf() força o sympy a devolver um número (e não uma fração, por exemplo)
#Verifique que a soma de rieman é uma boa aproximação para a integral!
```
